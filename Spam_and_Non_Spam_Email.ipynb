```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
# Logistic regresion
from sklearn.linear_model import LogisticRegression
```

```python
df=pd.read_csv('/content/drive/MyDrive/Dataset/spam.csv')
print(df.shape)
```

```python
X_train,X_test,y_train,y_test=train_test_split(df.Message,df.Category,test_size=0.20)
print(X_train.shape,y_train.shape)
print(X_test.shape,y_test.shape)
```

```python
count=0
for i in y_train:
  if i=='spam':
    count+=1
print(count)
```

```python
WC={}
for i in zip(X_train,y_train):
  if i[1]=='spam':
    for j in i[0].split():
      if j not in WC:
        WC[j]=1
      else:
        WC[j]+=1
print(WC)
SpamWord=[]
for i,j in zip(WC.keys(),WC.values()):
  if j>=count/12:
    SpamWord.append(i)
print(SpamWord)
```

```python
NumofWord=[0]*len(X_train)
for i in range(len(X_train)):
  for j in X_train.iloc[i].split():
    if j in SpamWord:
      NumofWord[i]+=1
A=np.sort(NumofWord)
print(A)
```

```python
model=LogisticRegression()
NumofWord_reshaped = np.array(NumofWord).reshape(-1, 1)
model.fit(NumofWord_reshaped,y_train)
```

```python
NumofWord=[0]*len(X_test)
for i in range(len(X_test)):
  for j in X_test.iloc[i].split():
    if j in SpamWord:
      NumofWord[i]+=1
```

```python
from collections import Counter
le=model.predict(np.array(NumofWord).reshape(-1,1))
print(Counter(le))
print(Counter(y_test))
accuracy = accuracy_score(y_test, le)

print(f"Accuracy: {accuracy*100}")
```
