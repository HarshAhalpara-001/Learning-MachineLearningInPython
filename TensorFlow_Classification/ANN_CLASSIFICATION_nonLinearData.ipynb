**Learning from this:** When dealing with datasets that exhibit linear patterns or overlap in classification, it is essential to explore alternative features. In such cases, especially when the dataset is known, we should employ nonlinear activation functions like ReLU, sigmoid, or tanh.

By default, if no activation function is specified in a neural network, it typically uses a linear activation function. This means that the output is simply a weighted sum of the inputs, which can limit the model's ability to capture complex relationships in the data.

For binary classification, I utilized the sigmoid function. As a result, the model's performance approached 98-100%, demonstrating that the data has become non-linearly separable, effectively forming a circular boundary as determined by the neural network.

```python
from sklearn import datasets
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
tf.__version__
```

```python
from sklearn import datasets
X,y=datasets.make_circles(n_samples=1000, shuffle=True, noise=0.03, random_state=42)
X.shape,y.shape,X,y
```

```python
plt.scatter(X[:,0], X[:,1], c=y)
```

```python
# model creation and creating a layes.
model_1=tf.keras.Sequential ([
    tf.keras.layers.Dense(100, activation='relu'),
    tf.keras.layers.Dense(10, activation="relu"),
    tf.keras.layers.Dense(1, activation="sigmoid")
])
# compilation of model
model_1.compile(loss=tf.keras.losses.BinaryCrossentropy(),
                optimizer=tf.keras.optimizers.Adam(),
                metrics=['accuracy'])
history=model_1.fit(X,y,epochs=100,verbose=1)
```

```python
import numpy as np

def plot_decision_boundary(model, X, y):
  """
  Plots the decision boundary created by a model predicting on X.
  This function has been adapted from two phenomenal resources:
   1. CS231n - https://cs231n.github.io/neural-networks-case-study/
   2. Made with ML basics - https://github.com/GokuMohandas/MadeWithML/blob/main/notebooks/08_Neural_Networks.ipynb
  """
  # Define the axis boundaries of the plot and create a meshgrid
  x_min, x_max = X[:, 0].min() - 0.1, X[:, 0].max() + 0.1
  y_min, y_max = X[:, 1].min() - 0.1, X[:, 1].max() + 0.1
  xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100),
                       np.linspace(y_min, y_max, 100))

  # Create X values (we're going to predict on all of these)
  x_in = np.c_[xx.ravel(), yy.ravel()] # stack 2D arrays together: https://numpy.org/devdocs/reference/generated/numpy.c_.html

  # Make predictions using the trained model
  y_pred = model.predict(x_in)

  # Check for multi-class
  if model.output_shape[-1] > 1: # checks the final dimension of the model's output shape, if this is > (greater than) 1, it's multi-class
    print("doing multiclass classification...")
    # We have to reshape our predictions to get them ready for plotting
    y_pred = np.argmax(y_pred, axis=1).reshape(xx.shape)
  else:
    print("doing binary classifcation...")
    y_pred = np.round(np.max(y_pred, axis=1)).reshape(xx.shape)

  # Plot decision boundary
  plt.contourf(xx, yy, y_pred, cmap=plt.cm.RdYlBu, alpha=0.7)
  plt.scatter(X[:, 0], X[:, 1], c=y, s=40, cmap=plt.cm.RdYlBu)
  plt.xlim(xx.min(), xx.max())
  plt.ylim(yy.min(), yy.max())
```

```python
plot_decision_boundary(model_1,X,y)
```
