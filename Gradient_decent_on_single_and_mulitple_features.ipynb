```python
lr=0.0000001
epoch=5000
theta=np.random.randn(14,1)
X_train_age= np.c_[X_train,np.ones((len(X_train)))]
y_train_age=np.array(y_train).reshape(-1,1)
print(theta,X_train_age)
```

```python
epoch=100000
for i in range(1,epoch+1):
  dtheta=2*X_train_age.T.dot(X_train_age.dot(theta)-y_train_age)/len(X_train)
  error=np.sum((X_train_age.dot(theta)-y_train_age)**2)
  theta=theta-lr*(dtheta)
  error=error/len(X_train)
  if i%10000==0:
    print(f"Slope or Weight is:{theta[:-1]}   intercept or bias is:{theta[-1]}")
    print(f"Error is:{error} in epoch no. {i}")
    print(" ")
# plt.scatter(X_train['age'],y_train_age)
plt.scatter(X_train['age'],X_train_age.dot(theta),color='red')
plt.show()
plt.plot(X_train['age'],y_train,'ro')
plt.xlabel('total_rooms')
plt.ylabel('median_house_value')
plt.show()
```
